This series of notebooks are used to develop a full NLP pipeline (tokenization, lemmatization, vectorization, cross-validation/testing) for creating a spam filter. The model will be trained/tested using a dataset available on Kaggle: https://www.kaggle.com/uciml/sms-spam-collection-dataset.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import string
import nltk
```

```python
df = pd.read_csv('spam.csv', encoding='latin-1')
pd.set_option('display.max_colwidth', 150) # Display 150 characters per text message
df.head()
```

Some texts 'spill' over onto multiple columns, so I decided to combine these into a new column. Also added column headings for the text message itself and the spam/ham label.

```python
df.drop(['v2', 'Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis=1, inplace=True)
df.columns = ['label', 'text']
```

## Exploratory Data Analysis

Exploring the dataset to see how many rows it has, number of spam/ham messages, average length of texts, etc. This will give me some insight regarding the full dataset before moving onto tokenization and data cleaning (e.g. removing stopwords/lemmatization).

```python
df.info()
```

```python
print('Dataset has {} rows and {} columns'.format(len(df), len(df.columns)))
print('There are {} messages labelled as spam, and {} messages labelled as ham'.format(len(df[df['label']=='spam']),
                                                                                       len(df[df['label']=='ham'])))

plt.figure()
sns.countplot(x="label", data=df)
plt.show()
```

```python
print('Average message length = {}'.format(round(df['text'].apply(lambda x: len(x) - x.count(' ')).mean(), 2))) # Doesn't count spaces
print('Average number of punctuation marks in message = {}'.format(round(df['text'].apply(lambda x: sum([1 for char in x if char in string.punctuation])).mean(), 2)))
```

```python
# Returns length of text message
def count_length(text):
    return round(len(text) - text.count(' '), 3)

# Returns ratio of punctuation marks in message
def count_punct(text):
    count = sum([1 for char in text if char in string.punctuation])
    ratio = count / count_length(text)
    return round(ratio, 3)

# Returns ratio of captial letters in message
def count_capital(text):
    count = sum([1 for char in text if char.isupper()])
    ratio = count / count_length(text)
    return round(ratio, 3)
```

```python
df['text_len'] = df['text'].apply(lambda x: count_length(x))
df['text_punct'] = df['text'].apply(lambda x: count_punct(x))
df['text_cap'] = df['text'].apply(lambda x: count_capital(x))

df.head()
```

```python
spams = df[df['label']=='spam']
hams = df[df['label']=='ham']

len_bins = np.linspace(0, 250, 40)

plt.figure()
sns.distplot(spams['text_len'], kde=False, bins=len_bins, norm_hist=True, label='spam')
sns.distplot(hams['text_len'], kde=False, bins=len_bins, norm_hist=True, label='ham')
plt.title('Distribution of text message lengths')
plt.legend()
plt.xlabel('')
plt.show()

punct_bins = np.linspace(0, 0.5, 40)

plt.figure()
sns.distplot(spams['text_punct'], kde=False, bins=punct_bins, norm_hist=True, label='spam')
sns.distplot(hams['text_punct'], kde=False, bins=punct_bins, norm_hist=True, label='ham')
plt.title('Distribution of text message punctuation ratios')
plt.legend()
plt.xlabel('')
plt.show()

cap_bins = np.linspace(0, 0.5, 40)

plt.figure()
sns.distplot(spams['text_cap'], kde=False, bins=cap_bins, norm_hist=True, label='spam')
sns.distplot(hams['text_cap'], kde=False, bins=cap_bins, norm_hist=True, label='ham')
plt.title('Distribution of text message capital letter ratios')
plt.legend()
plt.xlabel('')
plt.show()
```

This suggests that spam messages tend to be longer than non-spam messages, and generally contain a larger ratio of capital letters. The percentage of punctuation in each message does not seem to vary much between spam and ham messages. However, I will include all of these features in the dataset as they may help the ML model to learn the difference between spam and ham.

## Data cleaning

After some preliminary exploratory data analysis, the next step is data cleaning. This involves:
1. **removing punctuation**
2. **tokenization** - splitting the message up into a list of individual words
3. **removing stop words** - common words such as 'the', 'from', 'in', etc. are assumed to be filler words and do not contribute to the meaning/semantics of the message
4. **lemmatization** - grouping together inflected forms of a word so they can be analysed as a single term; I chose this method over stemming as it is generally more accurate, as stemming simply 'cuts down' words in a crude manner

Generally some of these processes are available through packages such as NLTK, but I will try to write most of the code myself to improve my understanding of each step.

```python
def remove_punct(text):
    new_text = ''.join([char for char in text if char not in string.punctuation])
    return new_text

df['text_nopunct'] = df['text'].apply(lambda x: remove_punct(x))

df.head()
```

```python
def tokenize(text):
    tokenized_list = re.split('\W+', text) # Use regular expressions to split string along non-alphanumeric characters
    return tokenized_list[:-1]

# Make all characters lower case since Python is case sensitive, reduces total number of terms
df['text_tokenized'] = df['text_nopunct'].apply(lambda x: tokenize(x.lower()))

df.head()
```

```python
# Common stopwords in the English language that don't add semantic value to a message
stopwords = nltk.corpus.stopwords.words('english')
stopwords[:10]
```

```python
def remove_stopwords(text_list):
    new_list = [word for word in text_list if word not in stopwords]
    return new_list

df['text_nostop'] = df['text_tokenized'].apply(lambda x: remove_stopwords(x))

df.head()
```

```python
wnl = nltk.WordNetLemmatizer() # Using WordNet lemmatizer avaialable from the NLTK package

def lemmatize_text(text_list):
    new_list = [wnl.lemmatize(word) for word in text_list]
    return new_list

df['text_lemmatized'] = df['text_nostop'].apply(lambda x: lemmatize_text(x))

df.head()
```

```python
# Combining all steps into one function
def clean_text(text):
    wnl = nltk.WordNetLemmatizer()
    stopwords = nltk.corpus.stopwords.words('english')
    text = ''.join([char.lower() for char in text if char not in string.punctuation])
    tokenized_list = re.split('\W+', text)
    text = [wnl.lemmatize(word) for word in tokenized_list if word not in stopwords]
    return text
```

```python
df['text_cleaned'] = df['text'].apply(lambda x: clean_text(x))

df.head()
```

```python
# Create cleaned dataset used for ML

data_cols = ['label', 'text_len', 'text_punct', 'text_cap', 'text_cleaned']
data = pd.DataFrame(df[data_cols])
data.head()
```
